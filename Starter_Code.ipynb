## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns=['EIN', 'NAME'])
application_df.head()
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
application_df["APPLICATION_TYPE"].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = ["T9","T13","T12","T2","T25","T14","T29","T15","T17"]

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure replacement was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
application_df["CLASSIFICATION"].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
application_df["CLASSIFICATION"].value_counts().loc[lambda x : x>1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = application_df["CLASSIFICATION"].value_counts().loc[lambda x : x<1000]
classifications_to_replace = classifications_to_replace.to_frame()
classifications_to_replace = classifications_to_replace.reset_index()
classifications_to_replace = classifications_to_replace["CLASSIFICATION"].to_list()

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure replacement was successful
application_df['CLASSIFICATION'].value_counts()

```

```python
application_df["ASK_AMT"].value_counts()
```

```python
application_df.sort_values("ASK_AMT", ascending=False)
```

```python
bins = [0, 5000, 10000, 20000, 30000, 50000, 100000, 500000, 1000000, 10000000, 10000000000]

labels = ["5,000", "6,000-10,000", "10000-20000", "20000-30000", "30000-50000", "50000-100000", "100,000-500,000", "500,000-1,000,000", "1,000,000-10,000,000", ">10,000,000"]

application_df["ASK_AMT"] = pd.cut(application_df["ASK_AMT"],
                                   bins = bins, labels = labels, include_lowest=True)

application_df.sort_values("ASK_AMT", ascending=False)
```

```python
application_df["ASK_AMT"].value_counts()
```

```python
application_df["STATUS"].value_counts()
```

```python
application_df = application_df.drop(columns=["STATUS"])
application_df.head()
```

```python
application_df["SPECIAL_CONSIDERATIONS"].value_counts()
```

```python
application_df = application_df.drop(columns=["SPECIAL_CONSIDERATIONS"])
application_df.head()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_df_dummies = pd.get_dummies(application_df, dtype=int)
application_df_dummies.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = application_df_dummies.IS_SUCCESSFUL.values
X = application_df_dummies.drop(columns="IS_SUCCESSFUL").values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=100, activation="relu", input_dim=49))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=80, activation="tanh"))

nn.add(tf.keras.layers.Dense(units=80, activation="leaky_relu"))

nn.add(tf.keras.layers.Dense(units=80, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=150)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
